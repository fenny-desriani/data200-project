```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
X_vec = np.array([1, 3])
Y_vec = np.array([2, 4])

# Computed theta
theta_hat = 1.4

# Predicted Y values (Ŷ)
Y_hat_vec = theta_hat * X_vec

# Residuals
residuals_vec = Y_vec - Y_hat_vec

# Create the figure
plt.figure(figsize=(6, 6))

# Set limits and grid
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.xticks(np.arange(0, 5, 1))
plt.yticks(np.arange(0, 6, 1))
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Plot the actual data points (blue dots)
plt.scatter(X_vec, Y_vec, color='blue', label="Actual Data Points (Y)")

# Plot the predicted values (red dots)
plt.scatter(X_vec, Y_hat_vec, color='red', label="Predicted Data Points (Ŷ)")

# Draw the regression line
X_line = np.linspace(0, 3.5, 100)
Y_line = theta_hat * X_line
plt.plot(X_line, Y_line, linestyle="--", color='red', label="Regression Line")

# Draw residuals (gray vertical lines from actual Y to predicted Ŷ)
for i in range(len(X_vec)):
    plt.plot([X_vec[i], X_vec[i]], [Y_vec[i], Y_hat_vec[i]], linestyle=':', color='gray')

# Labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Geometric Interpretation of Regression")
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
X_points = np.array([1, 3])
Y_points = np.array([2, 4])

# New regression equation: Y = X + 1 (to achieve zero MSE)
X_line = np.linspace(0, 4, 100)
Y_line = X_line + 1  # New perfect fit regression line

# Create the plot
plt.figure(figsize=(6, 6))

# Set limits and grid
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.xticks(np.arange(0, 5, 1))
plt.yticks(np.arange(0, 6, 1))
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Plot actual data points (blue dots)
plt.scatter(X_points, Y_points, color='blue', label="Actual Data Points (Y)")

# Draw the new regression line (solid red line, since it perfectly fits)
plt.plot(X_line, Y_line, linestyle="-", color='red', label="Perfect Fit Regression Line (Y = X + 1)")

# Labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Perfect Fit Regression Line (4f)")
plt.legend()

# Show the plot
plt.show()
```
